```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tools.tools import add_constant
from statsmodels.regression.linear_model import OLS
```

```python
err = np.random.normal (loc=0.0, scale=50.0, size=200) #nilai error
```

```python
sns.distplot(err, norm_hist=False, kde=True)
plt.show()
```

```python
x = np.linspace(start=0.5, stop=100, num=200) #nilai x
```

```python
x
```

```python
y = 3 + 10*x + err
```

```python
print(y[:10])
```

```python
sns.regplot(x,y, fit_reg=False)
plt.show()
```

```python
import pandas as pd
```

```python
data = pd.DataFrame({'y':y,'x':x})
data.head()
```

```python
data = add_constant(data)
data.head()
```

```python
model = OLS(data['y'],data[['const','x']])
result = model.fit()
```

```python
print(result.summary())
```
