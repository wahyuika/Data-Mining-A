```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("games.csv")
data.head()
```

```python
data.describe()
```

```python
np.sum(data.isnull())
```

```python
from scipy.stats import mode
data['umpire_2B']=data['umpire_2B'].fillna(data['umpire_2B']).dropna().mode()[0]
np.sum(data.isnull())
```

```python
from scipy import stats
z = np.abs(stats.zscore(data._get_numeric_data()))
print(z)
data = data[(z<3).all(axis=1)]
print(data.shape)
```

```python
data = pd.read_csv("pitches.csv")
data.head()
```

```python
data.describe()
```

```python
np.sum(data.isnull())
```

```python
from scipy.stats import mode

data['ax']=data['ax'].fillna(data['ax'].mean())
data['ay']=data['ay'].fillna(data['ay'].mean())
data['az']=data['az'].fillna(data['az'].mean())
data['break_angle']=data['break_angle'].fillna(data['break_angle'].mean())
data['break_length']=data['break_length'].fillna(data['break_length'].mean())
data['code']=data['code'].fillna(data['code']).dropna().mode()[0]
data['break_y']=data['break_y'].fillna(data['break_y'].mean())
data['end_speed']=data['end_speed'].fillna(data['end_speed'].mean())
data['nasty']=data['nasty'].fillna(data['nasty'].mean())
data['pfx_x']=data['pfx_x'].fillna(data['pfx_x'].mean())
data['pfx_z']=data['pfx_z'].fillna(data['pfx_z'].mean())
data['pitch_type']=data['pitch_type'].fillna(data['pitch_type']).dropna().mode()[0]
data['px']=data['px'].fillna(data['px'].mean())
data['pz']=data['pz'].fillna(data['pz'].mean())
data['spin_dir']=data['spin_dir'].fillna(data['spin_dir'].mean())
data['spin_rate']=data['spin_rate'].fillna(data['spin_rate'].mean())
data['start_speed']=data['start_speed'].fillna(data['start_speed'].mean())
data['sz_bot']=data['sz_bot'].fillna(data['sz_bot'].mean())
data['sz_top']=data['sz_top'].fillna(data['sz_top'].mean())
data['type_confidence']=data['type_confidence'].fillna(data['type_confidence'].mean())
data['vx0']=data['vx0'].fillna(data['vx0'].mean())
data['vy0']=data['vy0'].fillna(data['vy0'].mean())
data['vz0']=data['vz0'].fillna(data['vz0'].mean())
data['x0']=data['x0'].fillna(data['x0'].mean())
data['y0']=data['y0'].fillna(data['y0'].mean())
data['z0']=data['z0'].fillna(data['z0'].mean())
data['zone']=data['zone'].fillna(data['zone'].mean())
np.sum(data.isnull())
```

```python
from scipy import stats
z = np.abs(stats.zscore(data._get_numeric_data()))
print(z)
data = data[(z<3).all(axis=1)]
print(data.shape)
```
